```python
soup = BeautifulSoup(data, 'html.parser')
```

```python
netflix_data = pd.DataFrame(columns=["Date", "Open", "High", "Low", "Close", "Volume"])
```

```python
for row in soup.find("tbody").find_all('tr'):
    col = row.find_all("td")
    date = col[0].text
    Open = col[1].text
    high = col[2].text
    low = col[3].text
    close = col[4].text
    adj_close = col[5].text
    volume = col[6].text

    # Finally we append the data of each row to the table
    netflix_data = pd.concat([netflix_data,pd.DataFrame({"Date":[date], "Open":[Open], "High":[high], "Low":[low], "Close":[close], "Adj Close":[adj_close], "Volume":[volume]})], ignore_index=True)
```

```python
url = "https://cf-courses-data.s3.us.cloud-object-storage.appdomain.cloud/IBMDeveloperSkillsNetwork-PY0220EN-SkillsNetwork/labs/project/amazon_data_webpage.html"
html_data  = requests.get(url).text
```

```python
import pandas as pd
soup = BeautifulSoup(html_data, 'html.parser')

tesla_revenue = pd.DataFrame(columns=['Date', 'Revenue'])

for row in soup.find("tbody").find_all("tr"):
    col = row.find_all("td")
    date = col[0].text
    revenue = col[1].text.replace('$', '').replace(',', '')
    tesla_revenue = pd.concat([tesla_revenue, pd.DataFrame({'Date': [date], 'Revenue': [revenue]})], ignore_index=True)

tesla_revenue["Revenue"] = tesla_revenue['Revenue'].astype(float)

print(tesla_revenue.tail())
```

```python
url = "https://cf-courses-data.s3.us.cloud-object-storage.appdomain.cloud/IBMDeveloperSkillsNetwork-PY0220EN-SkillsNetwork/labs/project/gme_data_webpage.html"
html_data  = requests.get(url).text
```

```python
import pandas as pd
from bs4 import BeautifulSoup

soup = BeautifulSoup(html_data, 'html.parser')

# Find the table containing the revenue data
# I will look for a table with a relevant class or structure
# Based on typical Yahoo Finance pages, it might be a table with a specific class or within a specific div
# Let's try to find all tables and inspect their content to identify the correct one.

tables = soup.find_all('table')

# Assuming the revenue data is in one of the tables, iterate through them and check their content
gme_revenue = pd.DataFrame(columns=['Date', 'Revenue'])

for table in tables:
    if 'GameStop Quarterly Revenue' in str(table): # Look for a characteristic string in the table
        for row in table.find("tbody").find_all("tr"):
            col = row.find_all("td")
            if len(col) > 1: # Ensure there are enough columns to extract data
                date = col[0].text
                revenue = col[1].text.replace('$', '').replace(',', '')
                gme_revenue = pd.concat([gme_revenue, pd.DataFrame({'Date': [date], 'Revenue': [revenue]})], ignore_index=True)
        break # Stop after finding the correct table

# Clean the 'Revenue' column
gme_revenue["Revenue"] = gme_revenue['Revenue'].astype(float)

print(gme_revenue.tail())
```

```python
import matplotlib.pyplot as plt

def make_graph(stock_data, title):
    plt.figure(figsize=(10, 6))  # Optional: set figure size
    plt.plot(stock_data['Date'], stock_data['Open'])
    plt.xlabel("Date")
    plt.ylabel("Open Price")
    plt.title(title)
    plt.show()

make_graph(tesla_data, 'Tesla Stock Price Over Time')
```

```python
make_graph(gme_data, 'GameStop Stock Price Over Time')
```
